```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,root_mean_squared_error
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.Figure(figsize=(12,15))
sns.pairplot(df,x_vars=["YearsExperience"],y_vars=["Salary"],size=7,kind= 'scatter')
plt.xlabel("Years")
plt.ylabel("Salary")
plt.title("Salary prediction")
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x.head()
```

```python
y.head()
```

```python
# x = df["YearsExperience"]
# y = df["Salary"]
x_train,x_test,y_train,y_test = train_test_split(x ,y ,train_size = .8, test_size=.2, random_state=42 ,shuffle=True)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
plt.scatter(x_train ,y_train ,color = "red" )
plt.plot(x_train, model.predict(x_train),color = "blue")
plt.title("Salary vs Experience if years")
plt.xlabel("years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test ,y_test ,color = "red" )
plt.plot(x_test, model.predict(x_test),color = "blue")
plt.title("Salary vs Experience if years")
plt.xlabel("years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print(mse)
rms = root_mean_squared_error(y_test,y_pred)
print(rms)
r2score = r2_score(y_test,y_pred)
print(r2score)
```
