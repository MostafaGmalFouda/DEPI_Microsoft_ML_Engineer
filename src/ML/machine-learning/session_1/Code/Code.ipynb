```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4])
y = np.array([2,2.8,3.6,4.5])
```

```python
w=0
b=0
alpha = 0.01
epochs = 20
```

```python
sse_values = []

for i in range(epochs):
    y_hat = w*x+b

    D_w = 2/4 * np.sum((y_hat-y)*x)
    D_b = 2/4 * np.sum(y_hat-y)

    w -= alpha * D_w
    b -= alpha * D_b

    sse = ((np.sum(y_hat - y))**2)
    sse_values.append(sse)

    if(i+1) % 20 == 0:
        print(f"iteration {i+1}, sse{sse}")
```

```python
print(f"Optimixed param : slop = {w}, bias ={b}")
```

```python
plt.Figure(figsize=(20,15))
plt.subplot(1,2,1)
plt.plot(range(epochs),sse_values, label = "sse")
plt.xlabel("Iterations")
plt.ylabel("SSE")
plt.title("SSE over Iteration")
plt.legend()
plt.subplot(1,2,2)
plt.scatter(x ,y ,color = "red", label = "Data points")
plt.plot(x ,w*x+b ,color = "blue", label = "Regression line")
plt.xlabel("x")
plt.ylabel("Y")
plt.title("Linear Regression fit line")
plt.show()
```
