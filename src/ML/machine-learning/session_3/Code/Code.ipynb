```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x,y)
y_pred_lin = lin_reg.predict(x)
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_ridge = Ridge(alpha=10.0)
lin_ridge.fit(x,y)
y_pred_ridge = lin_ridge.predict(x)
print(lin_ridge.coef_)
print(lin_ridge.intercept_)
```

```python
lin_lasso = Lasso(alpha=10.0)
lin_lasso.fit(x,y)
y_pred_lasso = lin_lasso.predict(x)
print(lin_lasso.coef_)
print(lin_lasso.intercept_)
```

```python
print("Linear Regression")
print("Mse :",mean_squared_error(y,y_pred_lin))
print("r2_score :",r2_score(y,y_pred_lin))
print("="*50)

print("Ridge Regression")
print("Mse :",mean_squared_error(y,y_pred_ridge))
print("r2_score :",r2_score(y,y_pred_ridge))
print("="*50)

print("Lasso Regression")
print("Mse :",mean_squared_error(y,y_pred_lasso))
print("r2_score :",r2_score(y,y_pred_lasso))
print("="*50)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.scatter(x ,y ,color = "blue" ,label = "Actual data" )
plt.plot(x ,y_pred_lin ,color = "red" ,label = "Linear Regression")
plt.title("Linear Regression")
plt.legend()

plt.subplot(1,3,2)
plt.scatter(x ,y ,color = "blue" ,label = "Actual data" )
plt.plot(x ,y_pred_ridge ,color = "red" ,label = "Ridge Regression")
plt.title("Ridge Regression")
plt.legend()

plt.subplot(1,3,3)
plt.scatter(x ,y ,color = "blue" ,label = "Actual data" )
plt.plot(x ,y_pred_lasso ,color = "red" ,label = "Lasso Regression")
plt.title("Lasso Regression")
plt.legend()

plt.show()
```

```python
plt.scatter(x ,y ,color = "blue" ,label = "Actual data" )
plt.plot(x ,y_pred_lin ,color = "red" ,label = "Linear Regression")
plt.plot(x ,y_pred_ridge ,color = "green" ,linestyle = "--" ,label = "Ridge Regression")
plt.plot(x ,y_pred_lasso ,color = "purple" ,linestyle = ":" ,label = "Lasso Regression")
plt.title("Comparison of Linear vs Ridge vs Lasso Regression")
plt.legend()
plt.show()
```
