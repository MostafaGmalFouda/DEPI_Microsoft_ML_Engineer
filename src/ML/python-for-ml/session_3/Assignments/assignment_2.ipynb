```python
import os
import random
def make_folder(num_files=10):
    """ 
    This function creates a folder, generates files, deletes half of them,
    and prints the number of files before and after deletion.
    parameters:
    num_files (int): The number of files to create in the folder.
    Returns:
    list: list of files after deletion.
    Steps:
    1. Create a folder named 'assignment_2_folder'.
    2. Generate 'num_files' text files in the folder.
    3. Print the number of files before deletion.
    4. print the list of files before deletion.
    5. Randomly delete half of the files.
    6. Print the number of files after deletion
    7. return the list of files after deletion.
    """
    folder_name = "assignment_2_folder"
    os.makedirs(folder_name, exist_ok=True)

    for i in range(num_files):
        with open(os.path.join(folder_name, f"file_{i}.txt"), 'w') as f:
            f.write(f"This is file {i}\n")

    files = os.listdir(folder_name)
    print(f"Number of files before deletion: {len(files)}")
    Files_1 = []
    for file in files:
        Files_1.append(file)
    print(f"files before deletion: {Files_1}") 

    files_to_delete = random.sample(files, len(files) // 2)
    for file_name in files_to_delete:
        os.remove(os.path.join(folder_name, file_name))

    remaining_files = os.listdir(folder_name)
    print(f"Number of files after deletion: {len(remaining_files)}")
    Files_2 = []
    for file in remaining_files:
        Files_2.append(file)
    return f"files before deletion: {Files_2}"
    
```

```python
make_folder(10)
```
