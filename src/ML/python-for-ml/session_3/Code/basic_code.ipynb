```python
def user_info(name="Guest",age=25):
    """
    this function take name and age from user and return again
    : parem_1 : take the name from input user interface
    type parem_1 : str
    : parem_2 : take the age from input user interface
    type parem_2 : int

     return : this function will print the name and the age
     rtype : str,int

    """
    return f"my name is {name}, my age is {age}"
```
