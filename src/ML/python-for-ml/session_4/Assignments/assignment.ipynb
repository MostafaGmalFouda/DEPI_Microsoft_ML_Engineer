```python
class FIBONACCHI:
    """
    This class generates Fibonacci numbers up to a specified limit.\n
    : Attributes  
    limit (int): The number of Fibonacci numbers to generate.\n
    : Methods\n 
    fibonacchi(): Generates Fibonacci numbers up to the specified limit.
    Returns:
    list: A list containing the generated Fibonacci numbers.
    """
    def __init__(self,limit:int):
        self.limit = limit

    def fibonacchi(self):
        """
        This function generates Fibonacci numbers up to a specified limit.
        Parameters:
        limit (int): The number of Fibonacci numbers to generate. Default is 10.
        Returns:
        list: A list containing the generated Fibonacci numbers.\n
        Steps:
        1. Initialize the first two Fibonacci numbers.
        2. Use a loop to generate subsequent Fibonacci numbers.
        3. Print each Fibonacci number until the limit is reached.
        4. Return the list of generated Fibonacci numbers.
        """
        a, b = 0, 1
        fibonacci_numbers = []
        
        for i in range(self.limit):
            fibonacci_numbers.append(a)
            a, b = b, a + b
            i+=1
        
        return  f" Output: {fibonacci_numbers}"
    
f = FIBONACCHI(10)
f.fibonacchi()
```

```python
class REMOVE_ITEM:
    """
    This class removes an item from a given set.\n
    : Attributes\n
    set_items (set): The set from which to remove the item.\n
    item (str): The item to be removed from the set.\n
    : Methods\n
    remove_item(): Removes the specified item from the set.
    Returns:
    set: The updated set after removing the item or a message if the item was not found.
    """
    def __init__(self,set_items:set, item):
        self.set_items = set_items
        self.item = item

    def remove_item(self):
        """
        This function removes an item from a given set .
        parameters:
        set_items (set): The set from which to remove the item.
        item (str): The item to be removed from the set.
        Returns:
        set: The updated set after removing the item or a message if the item was not found.\n
        Steps:
        1. Check if the item exists in the set.
        2. If it exists, remove the item and return the updated set.
        3. If it does not exist, return a message indicating the item was not found.
        """
        if self.item in self.set_items:
            self.set_items.remove(self.item)
            return f"Updated set: {self.set_items}"
        else:
            return f"Item '{self.item}' not found in the set."
        
r = REMOVE_ITEM({1,2,3,4,5},3)
r.remove_item()
```

```python
class INDEX:
    """
    This class finds the index of a specified item in a given tuple.\n
    : Attributes\n
    tuple_items (tuple): The tuple in which to search for the item.\n
    item: The item whose index is to be found in the tuple.\n
    : Methods\n
    find_index(): Finds the index of the specified item in the tuple.
    Returns:
    int: The index of the item if found, otherwise a message indicating the item was not found.
    """
    def __init__(self,tuple_items:tuple,item):
        self.tuple_items = tuple_items
        self.item = item

    def find_index(self):
        """
        This function finds the index of a specified item in a given tuple.
        parameters:
        tuple_items (tuple): The tuple in which to search for the item.
        item: The item whose index is to be found in the tuple.
        Returns:
        int: The index of the item if found, otherwise a message indicating the item was not found.\n
        Steps:
        1. Use the index() method to find the index of the item in the tuple.
        2. If the item is found, return its index.
        3. If the item is not found, return a message indicating the item was not found.
        """
        if self.item in self.tuple_items:
            return f"index of {self.item} : {self.tuple_items.index(self.item)}"
        else:
            return f"item  {self.item} not found in the tuple. "
        
i = INDEX(('apple', 'banana', 'orange', 'grape', 'banana'),'orange')
i.find_index()
```

```python
class SPLIT:
    """ 
    This class splits a given sentence into individual words.\n
    : Attributes\n
    sentence (str): The sentence to be split into words.\n
    : Methods\n
    split_sentence(): Splits the sentence into individual words.
    Returns:
    list: A list of words obtained by splitting the sentence.
    """
    def __init__(self,sentence:str):
        self.sentence = sentence

    def split_sentence(self):
        """ 
        This function splits a given sentence into individual words.
        parameters:
        sentence (str): The sentence to be split into words.
        Returns:
        list: A list of words obtained by splitting the sentence.\n
        Steps:
        1. Use the split() method to divide the sentence into words based on spaces.
        2. Return the list of words.
        """
        return f"list of words in sentence: {self.sentence.split()}"
    
s = SPLIT("Welcome to Amit")
s.split_sentence()
```

```python
class CHECKPALINDROME:
    """
    This class checks if a given string is a palindrome.\n
    : Attributes\n
    string (str): The string to be checked for palindrome properties.\n 
    : Methods\n
    check_palindrome(): Checks if the string is a palindrome.
    Returns:
    str: "palindrome" if the string is a palindrome, "Not a palindrome" otherwise.
    """

    def __init__(self,string:str):
        self.string = string

    def check_palindrome(self):
        """ 
        This function checks if a given string is a palindrome.
        parameters:
        string (str): The string to be checked for palindrome properties.
        Returns:
        str: "palindrome" if the string is a palindrome, "Not a palindrome" otherwise.\n
        Steps:
        1. Reverse the string.
        2. Compare the reversed string with the original string.
        3. Return the appropriate message based on the comparison.
        
        """
        reverse_string = self.string[::-1]
        if reverse_string == self.string :
            return f"string '{self.string}' is palindrome."
        else:
            return f"string '{self.string}' is Not a palindrome."
        
c = CHECKPALINDROME("racecar")
c.check_palindrome()
```

```python
class FACTORIAL:
    """
    This class calculates the factorial of a given number.\n
    : Attributes\n
    number (int): The number for which to calculate the factorial.\n
    : Methods\n
    factorial(): Calculates the factorial of the specified number.
    Returns:
    int: The factorial of the given number.
    """

    def __init__(self,number:int):
        self.number = number

    def factorial(self):
        """
        This function calculates the factorial of a given number.
        parameters:
        number (int): The number for which to calculate the factorial.
        Returns:
        int: The factorial of the given number.\n
        Steps:
        1. Initialize a variable to hold the factorial result.
        2. Use a loop to multiply the numbers from 1 to the given number.
        """
        factorial = 1
        for i in range(1,self.number+1):
            factorial*=i
        return f"Factorial of number {self.number} is {factorial}"
    
fac = FACTORIAL(5)
fac.factorial()
```

```python
class SQUARES:
    """
    This class finds the squares of numbers from 1 to a given number.\n
    : Attributes\n
    number (int): The upper limit for finding squares.\n
    : Methods\n
    find_squares(): Finds the squares of numbers from 1 to the given number.
    Returns:
    list: A list of squares of numbers from 1 to the given number.
    """
    def __init__(self,number:int):
        self.number = number

    def find_squares(self):
        """
        This function finds the squares of numbers from 1 to the given number.
        parameters:
        number (int): The upper limit for finding squares.
        Returns:
        list: A list of squares of numbers from 1 to the given number.\n
        Steps:
        1. Initialize an empty list to hold the squares.
        2. Use a loop to calculate the square of each number from 1 to the given number.
        3. Append each square to the list.
        4. Return the list of squares.
        """
        squares = []
        for i in range(1,self.number+1):
            squares.append(i**2)
        return f"Squares of numbers from 1 to {self.number} are: {squares}"
    
sq = SQUARES(10)
sq.find_squares()
```

```python
class SUMEVEN:
    """
    This class calculates the sum of even numbers in a given list.\n
    : Attributes\n
    numbers (list): A list of numbers to check for evenness.\n
    : Methods\n
    sum_even(): Calculates the sum of even numbers in the list.
    Returns:
    int: The sum of even numbers in the list.
    """

    def __init__(self,numbers:list):
        self.numbers = numbers

    def sum_even(self):
        """
        This function calculates the sum of even numbers in a given list.
        parameters:
        numbers (list): A list of numbers to check for evenness.
        Returns:
        int: The sum of even numbers in the list.\n
        Steps:
        1. Initialize a variable to hold the sum of even numbers.
        2. Use a loop to iterate through each number in the input list.
        3. Check if each number is even, and if so, add it to the sum.
        4. Return the total sum of even numbers.
        """
        sum_even_numbers = 0
        for number in self.numbers:
            if number % 2 == 0:
                sum_even_numbers += number
        return f"Sum of even numbers: {sum_even_numbers}"
    
sum = SUMEVEN([1,2,3,4,5,6])
sum.sum_even()
```

```python
class REVERSE:
    """
    This class reverses a given list.\n
    : Attributes\n
    list (list): The list to be reversed.\n
    : Methods\n
    reverse_list(): Reverses the specified list.
    Returns:
    list: The reversed list.
    """

    def __init__(self,list:list):
        self.list = list

    def reverse_list(self):
        """ 
        This function reverses a given list.
        parameters:
        list (list): The list to be reversed.
        Returns:
        list: The reversed list.\n
        Steps:
        1. Use slicing to reverse the list.
        2. Return the reversed list.
        """
        return f"reversed list: {self.list[::-1]}"
    
re = REVERSE([1,2,3])
re.reverse_list()
```

```python
class MEDIAN:
    """
    
    """

    def __init__(self,list:list):
        self.list = list

    def find_median(self):
        """
        This function finds the median of a given list of numbers.
        parameters:
        list (list): The list of numbers to find the median from.
        Returns:
        float: The median value of the list.\n
        Steps:
        1. Sort the list of numbers.
        2. Determine if the length of the list is odd or even.
        3. Calculate and return the median based on the length.
        """
        sorted_list = sorted(self.list)
        length = len(sorted_list)
        mid = length // 2 # we use floor division to get the middle index because index start from 0
        
        # Check if the length of the list is even or odd if even we take average of two middle numbers else we take the middle number
        if length % 2 == 0:
            median = (sorted_list[mid - 1] + sorted_list[mid]) / 2
        else:
            median = sorted_list[mid]

        return f"Median of the list is: {median}"
m = MEDIAN([5,2,7,3,1])
m.find_median()
```

```python
class MODE:
    """
    
    """

    def __init__(self,list:list):
        self.list = list

    def find_mode(self):
        """
        This function finds the mode of a given list of numbers.
        parameters:
        list (list): The list of numbers to find the mode from.
        Returns:
        any: The mode value of the list. we choose any type because mode can be string or number.\n
        Steps:
        1. Create a dictionary to count the occurrences of each number.
        2. Iterate through the list and populate the dictionary with counts.
        3. Determine the number with the highest count (the mode).
        4. Return the mode value.
        """
        count_dict = {}
        for number in self.list:
            if number in count_dict:
                count_dict[number] += 1
            else:
                count_dict[number] = 1
        
        mode = max(count_dict, key=count_dict.get)
        return f"Mode of the list is: {mode}"
mod = MODE([5, 2, 7, 3, 2, 1, 5])
mod.find_mode()
```
