```python
def is_prime(num:int):
    """
    
    """
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
    
```

```python
is_prime(6)
```

```python
def reverse_text(text:str = "mostafa Gamal" ):
    split_text = text.split(" ")
    new_list = split_text[::-1]
    result = "".join(new_list)
    return result
```

```python
"""
this class

"""
class STUDENTINFO:
    # __ __ its name is dunderd 
    def __init__(self,student_name = "Mostafa",student_age = 20,GPA = 3.4,gender = "male"):
        self.student_name = student_name
        self.student_age = student_age
        self.GPA = GPA
        self.gender = gender

    def student_info(self):
        return f"Welcome into depi program, {self.student_name} your age is {self.student_age} and GPA is {self.GPA} and gender is {self.gender}"
```

```python
from abc import ABC,abstractclassmethod
class ANIMAL(ABC):
    @abstractclassmethod
    def move(self):
        pass
```

```python
# any class will inhert from abstract class it must be imblement all of abstract method on it 
class BIRD(ANIMAL):

    def move(self):
        pass
class CAT(ANIMAL):
    
    def move(self):
        print("move from cat")
```

```python
a = BIRD()
a.move()
```

```python
b = CAT()
b.move()
```
