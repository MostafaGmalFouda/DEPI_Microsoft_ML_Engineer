## Task_1

```python
email = "Amit_ml@gmail.edu"
# check if count of "@" is exactly one if true it is valid if false it is invalid
# after check of @ check if the "." is at least one after the "@" symbol if not it is invalid
# count() return the number of non-overlapping occurrences of substring sub in string
if email.count("@")== 1 :
    if email.count(".") >= 1:
        print("Valid Email")
else:
    print("Invalid Email")
```

```python
# Extract and return the part of the email before the "@" symbol for that we will use string slicing 
# we will use email[start index:end index : steps] note that end index not included
print(email[0:7:1])
```

```python
# Extract and return the domain (the part between "@" and the last ".") for that we will use string slicing 
# we will use email[start index:end index : steps] note that end index not included
print(email[8:13:1])
```

```python
# Check if the email ends with ".com". If it does, return "Commercial Domain".
#  If it ends with ".edu", return "Educational Domain".
#  Otherwise, return "Other Domain".
# we will use function endwith() return True if S ends with the specified suffix, False otherwise
if email.endswith(".com") == True:
    print("Commercial Domain")
elif email.endswith(".edu")== True:
    print("Educational Domain")
else:
    print("Other Domain")
```

## Task_2

```python
message = "###!!@mocleW EPGTQ!!!6789"
# Encoded Message: ###!!@mocleW EPGTQ!!!6789
# Extract the core part of the message: "mocleW EPGTQ" for that we will use string slicing 
# we will use message[start index:end index : steps] note that end index not included
core = message[6:18:1]
"""""
I coud use this instead of core = message[6:18:1] and that give me same result.
# function removeprefix() return a str with the given prefix string removed if present.
# function removesuffix() return a str with the given suffix string removed if present.
core = message.removeprefix("###!!@")
core = core.removesuffix("!!!6789")
"""
print(f"core of message is \"{core}\"")
# Reverse the first word: "mocleW" becomes "Welcome". for that we will use string slicing to get first word 
# we will use core[start index:end index : steps] note that end index not included 
# we use -step to get string form end
first_word = core[0:6:1]
first_word= first_word[::-1]
print(f"first word after reverse is \"{first_word}\"")
# Replace shifted vowels in the second word for that we will use string slicing to get first word
# we will use core[start index:end index : steps] note that end index not included
# after that we will use function replace() return a copy with all occurrences of substring old replaced by new
second_word = core[7::1]
second_word = second_word.replace("E","")
print(f"second word after replace is \"{second_word}\"")
# Final decoded message
print(f"Final decoded message is \"{first_word} {second_word}\"")
```
